```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
##lineal function

def f(x):
    m=7
    b=3
    return m*x+b

N=100

x=np.linspace(-10,10,N)

y=f(x)

fig, axes=plt.subplots()
axes.plot(x,y)
axes.grid()
```

```python
##polynomial function

def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x, y)
plt.grid()
```

```python
##trigonometric function

def f(x):
    return np.tan(x)

y=f(x)

plt.plot(x, y)
plt.grid()
```

```python
##exponential function

def f(x):
    return np.exp(x)

y=f(x)

plt.plot(x, y)
plt.grid()
```

```python
##logarithmic function

def f(x):
    return np.log(x)

x=np.linspace(0.0000001,10,N)

y=f(x)

plt.plot(x, y)
```

```python
##Heaviside function

def H(x):
    Y=np.zeros(len(x))
    for i in range(len(x)):
        if x[i]>0:
            Y[i]=1
        else:
            Y[i]=0
    return Y

N=1000
x=np.linspace(-10,10,N)
y=H(x)

plt.plot(x,y)
```

```python
##abs function

def f(x):
    return np.abs(x)

y=f(x)

plt.plot(x, y)
```

```python
##compound function

def g(x):
    return (x**2)

def f(x):
    return 7*x**3-2*x**2+x-3

y=f(x)

y_2=g(x)

y_f_g=f(g(x))

plt.plot(x, y)
plt.plot(x, y_2)
plt.plot(x, y_f_g)
```
